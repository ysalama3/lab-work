```python
#read in the file and make dictionaries

in_file = open("/home/eeb177-student/Desktop/eeb-177/lab-work/exercise-7/cocoli.txt")
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}
in_file.readline()
for line in in_file:
    items = line.split()
    tree_id = items[0]
    s94 = float(items[4])
    s97 = float(items[5])
    s98 = float(items[6])
    
    sizes_94[tree_id] = s94
    sizes_97[tree_id] = s97
    sizes_98[tree_id] = s98

    
in_file.close()
```
